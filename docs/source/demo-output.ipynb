# EK80 ADCP: Quick-start demo

This notebook shows how to load one or more EK80 ADCP NetCDF files and produce the standard diagnostic plots.

**Prerequisites**: install the package (`pip install -e .`) and activate its virtual environment.

```python
from pathlib import Path

from ek80adcp import logger, plotters
from ek80adcp.readers import find_ek80_files, load_ek80

logger.disable_logging()  # remove to see progress messages
```

## Load data

Point `data_dir` at the folder containing your raw EK80 NetCDF files (`DSMIXSED2-D*-T*.nc`), or pass an explicit `file_list` to `load_ek80()`.

```python
# Single file from the package test data
data_dir = Path("../../data")
files = find_ek80_files(str(data_dir))

ds = load_ek80(file_list=files)
print(ds)
```

## Inspect variables

```python
plotters.show_variables(ds)
```

## Time series at a single depth

```python
# Eastward and northward velocity at ~15 m depth (index 3)
fig, ax = plotters.plot_velocity_at_depth(ds, depth_idx=3)
```

## Hovmöller (depth vs time)

```python
# 3-panel Hovmöller: vx, vy, vz
fig, axes = plotters.plot_hovmoller(ds)
```

## Track with velocity vectors

```python
# Ship track with horizontal current arrows at the shallowest bin
fig, ax = plotters.plot_track_quiver(ds, depth_idx=0)
```
