## Demonstrator loading EK80 data

```python
from ek80adcp import logger, plotters
from ek80adcp.readers import load_ek80

logger.disable_logging()  # remove this line if you want log output
```

```python
# Load all EK80 ADCP files from data/
# Pass file_list=[...] to load specific files instead
ds = load_ek80()
print(ds)
```

```python
# Browse variables and their units
plotters.show_variables(ds)
```

```python
# Eastward and northward velocity at one depth bin (index 3 ≈ 14.7 m)
fig, ax = plotters.plot_velocity_at_depth(ds, depth_idx=3)
```

```python
# Depth–time Hovmöller for all three velocity components
fig, axes = plotters.plot_hovmoller(ds)
```

```python
# Depth–time Hovmöller for all three velocity components
fig, axes = plotters.plot_hovmoller(ds)
```

```python
# Platform track with current vector arrows (shallowest depth bin)
fig, ax = plotters.plot_track_quiver(ds, depth_idx=0)
```
